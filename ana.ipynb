````python
# 3D渲染环境测试，notebook中需要套件，可直接在浏览器中打开

import math
from pyecharts import options as opts
from pyecharts.charts import Line3D
from pyecharts.faker import Faker
from pyecharts.globals import CurrentConfig, NotebookType

# 确保 Pyecharts 在 Jupyter Notebook 中正确显示
CurrentConfig.NOTEBOOK_TYPE = NotebookType.JUPYTER_LAB

data = []
for t in range(0, 25000):
    _t = t / 1000
    x = (1 + 0.25 * math.cos(75 * _t)) * math.cos(_t)
    y = (1 + 0.25 * math.cos(75 * _t)) * math.sin(_t)
    z = _t + 2.0 * math.sin(75 * _t)
    data.append([x, y, z])

c = (
    Line3D()
    .add(
        "",
        data,
        xaxis3d_opts=opts.Axis3DOpts(Faker.clock, type_="value"),
        yaxis3d_opts=opts.Axis3DOpts(Faker.week_en, type_="value"),
        grid3d_opts=opts.Grid3DOpts(
            width=100, depth=100, rotate_speed=150, is_rotate=True
        ),
    )
    .set_global_opts(
        visualmap_opts=opts.VisualMapOpts(
            max_=30, min_=0, range_color=Faker.visual_color
        ),
        title_opts=opts.TitleOpts(title="Line3D-旋转的弹簧"),
    )
)

# 在网页中显示，打开浏览器即可
c.render("line3d.html")


````

````python
from pyecharts import options as opts
from pyecharts.charts import Map3D
from pyecharts.globals import GeoType, ChartType, ThemeType
from pyecharts.commons.utils import JsCode

# 数据准备
data_all = {
    "city": [
        # "鄂州", "宜昌", "荆州", "武汉", "襄阳", "孝感", "黄冈", "恩施", "荆门",
        # "黄石", "神农架", "随州", "十堰", "咸宁"
        "鄂州市", "宜昌市", "荆州市", "武汉市", "襄阳市", "孝感市", "黄冈市", "恩施市", "荆门市",
        "黄石市", "神农架林区", "随州市", "十堰市", "咸宁市"
    ],
    "total_score": [
        83.04798647, 80.44296421, 80.26584916, 79.29385101, 79.23896166,
        78.75694461, 78.13142456, 76.00076889, 75.67097309, 74.10479364,
        73.39117971, 72.01220165, 71.87595035, 71.24698471
    ]
}

data_pairs = list(zip(data_all["city"], data_all["total_score"]))

# 绘制 3D 地图
map3d = (
    Map3D(init_opts=opts.InitOpts(bg_color="#f5f5f5", theme=ThemeType.LIGHT))
    .add_schema(
        maptype="湖北",
        itemstyle_opts=opts.ItemStyleOpts(
            color="rgb(0,0,0)",
            opacity=1,
            border_width=0.8,
            border_color="rgb(62,215,213)"
        ),
        map3d_label=opts.Map3DLabelOpts(is_show=False),
        emphasis_label_opts=opts.LabelOpts(
            is_show=True,
            formatter=JsCode(
                "function(params){return params.name + ' : ' + params.value[2];}"
            ),
        ),
        light_opts=opts.Map3DLightOpts(
            main_color="#fff",
            main_intensity=1.5,
            is_main_shadow=False,
            main_alpha=55,
            main_beta=10,
            ambient_intensity=0.3,
        ),
    )
    .add(
        series_name="总分",
        data_pair=data_pairs,
        type_=ChartType.BAR3D,
        bar_size=1,
        shading="lambert",
        label_opts=opts.LabelOpts(
            is_show=True,
            formatter=JsCode(
                "function(params){return params.name + ' : ' + params.value[2];}"
            ),
        ),
        itemstyle_opts=opts.ItemStyleOpts(opacity=0.8),
    )
    .set_global_opts(
        title_opts=opts.TitleOpts(title="湖北省各市总分3D地图"),
        visualmap_opts=opts.VisualMapOpts(
            is_show=True,
            dimension=2,
            max_=100,
            min_=70,
            range_text=["High", "Low"],
            # in_range={"color": ["#50a3ba", "#eac763", "#d94e5d"]},
        ),
    )
)

# 渲染为 Notebook 中显示的 HTML
map3d.render("map3d.html")
````
